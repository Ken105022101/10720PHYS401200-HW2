## Monte Carlo simulation of the 2D Ising model

Write a program/function to perform a Monte Carlo simulation of the 2D Ising model in zero magnetic field.

$$
  H = -J \sum_{\langle i,j\rangle} \sigma_i \sigma_j,
$$

where $\sigma=\pm 1$ and the sum is over nearest-neighbor pairs of sites on a square lattice of $N = L \times L$ sites.

You should implement two different update schemes:
* Sequential update.
* Random update.

For both schemes, we define a Monte Carlo sweep as $N=L\times L$ local spin updates(flip attemps). At the end each Monte Carlo sweep, i.e., after $N$ local spin updates(flip attemps), you should measure the following quantities:
* Total energy $E$ and $E^2$.
* Total matnetization $M$, $|M|$, and $M^2$.

At the end of the simulation, calculate following quantities:
* Specific heat $C_v$
$$
  C_v \equiv \frac{\partial \langle E\rangle}{\partial T} = \frac{1}{k_B T^2} \left( \langle E^2\rangle -\langle E\rangle^2 \right).
$$
* Susceptbility of $M$
$$ 
  \chi_M \equiv \frac{\partial \langle M\rangle}{\partial h} = \frac{1}{k_B T} \left( \langle M^2\rangle -\langle M\rangle^2 \right).
$$
* Susceptbility of $|M|$
$$ 
  \chi_{|M|} \equiv \frac{\partial \langle |M|\rangle}{\partial h} = \frac{1}{k_B T} \left( \langle M^2\rangle -\langle |M|\rangle^2 \right).
$$

## Problem 1: $2\times 2$ lattice

The exact results for $2\times 2$ lattices are (cf. arXiv:0803.0217):
$$
  Z=2 e^{8\beta J} + 12 + 2 e^{-8\beta J}.
$$

$$ 
  \langle E\rangle = -\frac{1}{Z} \left[ 2(8)e^{8\beta J} + 2(-8)e^{-8\beta J} \right].
$$

$$ 
  \langle E^2\rangle = \frac{1}{Z} \left[ 2(64)e^{8\beta J} + 2(64)e^{-8\beta J} \right].
$$

$$ 
  \langle |M|\rangle = \frac{1}{Z} \left[ 2(4)e^{8\beta J} + 8(2) \right].
$$

$$ 
  \langle M^2\rangle = \frac{1}{Z} \left[ 2(16)e^{8\beta J} + 8(4) \right].
$$

* Plot specific heat per site $C_v/N$ vs temperature using exat result, MC (sequential update), MC (random update).
* Plot susceptbility of $|M|$ per site $\chi_{|M|}/N$ vs temperature using exat result, MC (sequential update), MC (random update).

You should perform at least $10^3$ Monte Carlo sweeps.

```python
import numpy as np
import matplotlib.pyplot as plt
# initialize system parameters
J = 1.0
kB = 1.0

# Hamiltonian
def H(s, L):
    E = 0
    for i in range(L):
        for j in range(L):
            E += s[i, j] * (s[(i + 1) % L, j] + s[i, (j + 1) % L])
    energy = -J * E
    return energy

# magnetization
def mag(s, L):
    M = 0.0
    for i in range(L):
        for j in range(L):
            M += s[i, j]
    return M

def random_spins(L):
    s = np.zeros([L, L], int)
    for i in range(L):
        for j in range(L):
            s[i, j] = 2 * int( 2 * np.random.random() ) -1
    return s

def dE(s, L, i, j):
    NB = s[(i + 1) % L, j]+s[i, (j + 1) % L]+s[(i - 1) % L, j]+s[i, (j - 1) % L]
    dE = 2 * s[i, j] * NB
    return dE

def printSPINS(s, L):
    print("  " + str(np.arange(0, L, 1)).strip('[ ]'))
    for i in range(L):
        row = str(i) + " "
        for j in range(L):
            if s[i, j] == +1: row += '＋'
            if s[i, j] == -1: row += '－'
        print(row)
    print()
def changeSpin(T, s, L):
    for i in range(L):
        for j in range(L):
            deltaE = dE(s, L, i, j)
            if deltaE > 0:
                B = np.exp(- deltaE / (kB * T))
                if B >= np.random.random():
                    s[i, j] = -s[i, j]
            else :
                s[i, j] = -s[i, j]
                    
        
def randchangeSpin(T, s, L):
    i = np.random.randint(0, L)
    j = np.random.randint(0, L)
    deltaE = dE(s, L, i, j)
    if deltaE > 0:
        B = np.exp(- deltaE / (kB * T))
        if B >= np.random.random():
            s[i, j] = -s[i, j]
    else :
        s[i, j] = -s[i, j]

# Problem 1
L = 2
spins1 = np.ones((L, L))
spins1r = np.ones((L, L))


printSPINS(spins, L)

Cvps = []
Cvpsr = []
Mps = []
Mpsr = []
for t in np.linspace(0.01, 5, 1000):
    Etotal = 0
    Estotal =0
    Etotalr = 0
    Estotalr =0
    mtotal = 0
    mstotal = 0
    mtotalr = 0
    mstotalr = 0
    for _ in range(1000):
        changeSpin(t, spins1, L)
        randchangeSpin(t, spins1r, L)
        Etotal += H(spins1, L)
        Estotal += H(spins1, L) * H(spins1, L)
        Etotalr += H(spins1r, L)
        Estotalr += H(spins1r, L) * H(spins1r, L)
        mtotal += abs(mag(spins1, L))
        mstotal += mag(spins1, L) * mag(spins1, L)
        mtotalr += abs(mag(spins1r, L))
        mstotalr += mag(spins1r, L) * mag(spins1r, L)
    Cvps.append(((Estotal / 1000) - (Etotal / 1000) ** 2) / (kB * (t ** 2) * (L ** 2)))
    Cvpsr.append(((Estotalr / 1000) - (Etotalr / 1000) ** 2) / (kB * (t ** 2) * (L ** 2)))
    Mps.append(((mstotal / 1000) - (mtotal / 1000) ** 2) / (kB * (L ** 2) * t))
    Mpsr.append(((mstotalr / 1000) - (mtotalr / 1000) ** 2) / (kB * (L ** 2) * t))
plt.subplot(211)
plt.plot(np.linspace(0.01, 5, 1000), Cvps, label = "Cv/N sequential")
plt.plot(np.linspace(0.01, 5, 1000), Cvpsr, label = "Cv/N random")
plt.legend(loc = 'best')
plt.subplot(212)
plt.plot(np.linspace(0.01, 5, 1000), Mps, label = "|M|/N sequential")
plt.plot(np.linspace(0.01, 5, 1000), Mpsr, label = "|M|/N random")
plt.legend(loc = 'best')
plt.show()
```
